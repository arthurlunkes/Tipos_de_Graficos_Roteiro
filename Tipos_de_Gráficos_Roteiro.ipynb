```python
df = pd.read_csv('vendas_produtos.csv')
df
```

```python
def gerarGrafico(style):
  #estilo de cor
  plt.style.use(style) 

  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico
  plt.show()

gerarGrafico('ggplot')
```

```python
df = pd.read_csv('Idade_Carro.csv')
df
```

```python
def gerarGraficoDispersao(style):
  # Criar o gráfico de dispersão
  plt.style.use(style)
  plt.scatter(df['Idade'], df['Preço'])

  plt.title('Preço do Carro x Idade')
  plt.xlabel('Idade do Carro (Anos)')
  plt.ylabel('Preço (R$)')

  # Mostrar o gráfico
  plt.show()

gerarGraficoDispersao('bmh')
```

# Explique como esse gráfico:

## Modifique a distribuição aleatória para distribuição normal - Pesquise a função numpy

```python
# Vamos gerar um conjunto de dados aleatórios para as idades
np.random.seed(0)
#idades = np.random.randint(0, 100, size=1000) #distribuição aleatória de valor inteiro

idades = np.random.normal(0, 100, size=1000) #distribuição normal de valor inteiro

# Criar o histograma
plt.hist(idades, bins=20, edgecolor='black')

plt.title('Distribuição das Idades')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Mostrar o gráfico
plt.show()
```

# Interprete a distribuição distribuição e compare a diferença com a distribuição normal

R: O gráfico contém idades negativas, e a frequência é maior quanto menor for a idade.

Uma distribuição aleatória refere-se a uma coleção de valores que são gerados de forma imprevisível, sem seguir um padrão específico. uma distribuição normal segue uma curva em forma de sino e tem uma maior concentração de valores em torno da média.

# Gráfico de Donut

```python
# Dados
generos = ['Feminino', 'Masculino']
quantidade = [650, 550]
```

```python
plt.figure(figsize=(7, 15), facecolor='#212946')

pie = plt.pie(quantidade, labels=generos, autopct='%1.1f%%', startangle=84, 
              colors=['#FFC0CB', '#0000FF'], wedgeprops={'width': 0.3}, 
              pctdistance=1.3, radius=0.7, labeldistance=1.5)

plt.title('Distribuição de Gênero', color= 'white', fontsize=30)

texts = pie[1]
texts[generos.index('Feminino')].set_fontsize(20)
texts[generos.index('Feminino')].set_color('white')
texts[generos.index('Masculino')].set_fontsize(20)
texts[generos.index('Masculino')].set_color('white')
texts = pie[2]
texts[generos.index('Feminino')].set_fontsize(20)
texts[generos.index('Feminino')].set_color('white')
texts[generos.index('Masculino')].set_fontsize(20)
texts[generos.index('Masculino')].set_color('white')
plt.show()
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: é uma maneira de visualizar a distribuição dos dados através de seus quartis. Ele mostra a mediana, o primeiro e terceiro quartis e os valores mínimo e máximo. O gráfico é rotulado com um título, rótulos de eixo x e y e é exibido usando plt.show().

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R: tem 2 pontas de extremidade dos pontos de cada equipe, e onde faz a onda é onde tem mais registros próximos daquele valor.
A equipe C tem pessoas com pontos que variam de usn 50 até 120, e tem mais pessoas com 90 pontos. enquanto a equipe D tem pessoas com pontos que variam de 70 até 110, e tem mais pessoas com 90 pontos.
